```python
import pandas as pd
import psycopg2
```

```python
df = pd.read_csv('wine.data', header=None)
df.head()
```

```python
try:
    conn = psycopg2.connect("dbname='postgres' user='postgres' host='db' password='1111'")
except Exception as e:
    print(f'I am unable to connect to the database:\n{e}')
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:1111@db:5432/postgres')
```

```python
df.to_sql('wine', engine)
```

```python
df_pg = pd.read_sql_query('select * from wine', con = engine)
df_pg.head(2)
```

```python
# drop table
cursor = conn.cursor()
cursor.execute('drop table wine')
cursor.execute('commit')
```
